## To fill/drop empty cells,the fillna(),dropna() function can be used

```python
data['Urban_Population'].unique()
```

```python
Urban_Population_mean=data['Urban_Population'].mean()
Urban_Population_mean
```

```python
Urban_Population_mean=data['Urban_Population'].median()
Urban_Population_mean
```

```python
data['Urban_Population'].fillna('Urban_Population_median',inplace=True)
```

```python
data.info()
```

```python
data['Literacy_Rate'].unique()
```

```python
Literacy_Rate_mean=data['Literacy_Rate'].mean()
Literacy_Rate_mean
```

```python
Literacy_Rate_median=data['Literacy_Rate'].median()
Literacy_Rate_median
```

```python
data['Literacy_Rate'].fillna(Literacy_Rate_median,inplace=True)
```

```python
data.info()
```

```python
data['Unemployment_Rate'].fillna(data['Unemployment_Rate'].median(),inplace=True)
```

```python
data.info()
```

```python
df=data
```

# Univeriate Analysis

**Numerical columns-Descriptive statistics**

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Population'],kde=True)
plt.title('Distribution of Population')
plt.show()
```

## boxplot

```python
sns.boxplot(x=df['Literacy_Rate'])
plt.title('Boxplot of Literacy Rate')
plt.show()
```

```python
df.info()
```

```python
df['Urban_Population']=pd.to_numeric(df['Urban_Population'],errors="coerce")
```

```python
plt.figure(figsize=(10,10))
df.boxplot()
```

## categorical columns

```python
sns.countplot(x='State',data=df)
plt.xticks(rotation=90)
plt.title('Frequency of records per state')
plt.show()
```

```python
df['Date'].value_counts().sort_index().plot(kind='line')
plt.title('Record over Time')
plt.show()
```

## Bivariate Analysis(2 variables)

Used to find relationships or associations between two variables.

### **Numerical vs Numerical**

```python
sns.scatterplot(x='Literacy_Rate',y='Unemployment_Rate',data=df)
plt.title('Literacy_Rate vs Unemployment_Rate')
plt.show()
```

```python
correlation=df.corr(numeric_only=True)
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title('correlation Matrix')
plt.show()
```

### categorical vs Numerical

```python
sns.boxplot(x='State',y='Population',data=df)
plt.xticks(rotation=90)
plt.title('population Distribution by state')
plt.show()
```

### Categorical VS Categorical

```python
pd.crosstab(df['State'],df['District'])
```

### Multivariate Analysis(3+ variables)

**pairplot**

Shows pairwise relationships and distributions

```python
sns.pairplot(df[['Population','Urban_Population','Rural_Population','Literacy_Rate','Unemployment_Rate']])
plt.show()
```

## Grouped Analysis
   

```python
df.groupby('State')[['Literacy_Rate','Unemployment_Rate']].mean().plot(kind='bar')
plt.title('Avg Literacy & Unemployment by state')
plt.ylabel('Rate')
plt.show()
```

### Multivariate Heatmap

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='viridis')
```

### Correlation Analysis in EDA


Correlation analysis is used to quantify the strength and direction of relationships between numerical variables. It helps identify how changes in one variable are associated with changes in another.

```python
correlation_Matrix=df.corr(numeric_only=True)
print(correlation_Matrix)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_Matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('correlation_Matrix Heatmap')
plt.show()
```
